```python
##coT 
# 解释什么是COT 
'''
CoT（思维链）是一种提示词技术，让大模型在输出最终答案之前，先展示一步步的推理过程，而不是直接给出结果。简单说就是让模型"把思考过程写出来"。

原理

# ❌ 普通提示（直接回答）
Q: 小明有5个苹果，给了小红2个，又买了3个，现在有几个？
A: 6个

# ✅ CoT 思维链提示（展示推理步骤）
Q: 小明有5个苹果，给了小红2个，又买了3个，现在有几个？
A: 
小明最初有5个苹果。
给了小红2个：5 - 2 = 3个。
又买了3个：3 + 3 = 6个。
所以小明现在有6个苹果。
为什么有效
复杂问题需要多步推理时，模型直接跳到最后答案容易出错
让模型"一步一步想"，相当于给了它一张草稿纸，能显著提升准确率
尤其擅长数学题、逻辑推理、常识推理等任务
常见的 CoT 变体
类型	做法	示例提示词
Zero-shot CoT	直接加一句"Let's think step by step"	"请逐步思考"
Few-shot CoT	给几个带推理过程的示例	给2-3个例子，每个都包含推理步骤
Self-Consistency CoT	多次运行取多数答案	跑5次CoT，选出现最多的结果
实际应用
比如在优化提示词这个场景，加上 CoT 可以让模型分析得更全面：


# 零样本 CoT 的写法
prompt = "请帮我优化以下提示词，<问题内容>，请逐步分析并给出建议。"
简单来说：CoT = 让模型把脑子里的草稿写出来，特别适合需要推理的复杂任务。
'''

instructions = """
给定一段用户与手机流量套餐客服的对话，。
你的任务是判断客服的回答是否符合下面的规范：

- 必须有礼貌
- 必须用官方口吻，不能使用网络用语
- 介绍套餐时，必须准确提及产品名称、月费价格和月流量总量。上述信息缺失一项或多项，或信息与事实不符，都算信息不准确
- 不可以是话题终结者

已知产品包括：

经济套餐：月费50元，月流量10G
畅游套餐：月费180元，月流量100G
无限套餐：月费300元，月流量1000G
校园套餐：月费150元，月流量200G，限在校学生办理
"""

# 输出描述
output_format = """
如果符合规范，输出：Y
如果不符合规范，输出：N
"""


context = """
用户：你们有什么流量大的套餐
客服：亲，我们现在正在推广无限套餐，每月300元就可以享受1000G流量，您感兴趣吗？
"""

cot = ""
# cot = "请一步一步分析对话"

'''<think>我们分析客服的回答。用户问流量大的套餐，客服回答："亲，我们现在正在推广无限套餐，每月300元就可以享受1000G流量，您感兴趣吗？"

检查规范：1. 礼貌？用了"亲"，虽然网络用语，但"亲"是电商常用，但规范说不能使用网络用语？"亲"属于网络用语，不规范。2. 官方口吻？用了"亲"不够官方。3. 介绍套餐时：准确提及产品名称（无限套餐）、月费价格（300元）、月流量总量（1000G），信息准确。4. 不是话题终结者？最后问"您感兴趣吗？"是引导后续对话，不是终结。

但第一条"必须有礼貌"，第二条"必须用官方口吻，不能使用网络用语"。"亲"是网络用语，不符合官方口吻。所以不符合规范。

因此输出N。</think>

不符合规范。客服使用了网络用语“亲”，不是官方口吻，因此违反规范要求。

N
'''

user_prompt = f"""
# 输出格式
{output_format}
{cot}

# 输入
{context}
"""

messages = [
    {"role": "system", "content": instructions},
    {"role": "user", "content": user_prompt},
]
response = get_message(messages,model="deepseek-ai/DeepSeek-V4-Flash",isStream= True, temperature=0.5)
print_response(response)
```
