## 1.文本模型

```python
from openai import OpenAI

# 创建 OpenAI 兼容客户端，指向 Agnes AI
client = OpenAI(
    api_key="YOUR_API_KEY",  # 替换为你的 API Key
    base_url="https://apihub.agnes-ai.com/v1"
)

# 调用 Agnes 2.0 Flash 模型
response = client.chat.completions.create(
    model="agnes-2.0-flash",
    messages=[
        {"role": "system", "content": "You are a helpful AI assistant."},
        {"role": "user", "content": "Explain how autonomous agents use tools to complete tasks."}
    ],
    temperature=0.7,
    max_tokens=1024
)

# 输出结果
print(response.choices[0].message.content)
```

### （1)加载环境变量

```python
import os

api_key=os.getenv('AGNES_API_KEY')
# print(api_key)
```

### (2)使用 requests 库调用

```python
import requests
import json

url = "https://apihub.agnes-ai.com/v1/chat/completions"
headers = {
    "Authorization": f"Bearer {api_key}",  # 替换为你的 API Key
    "Content-Type": "application/json"
}
payload = {
    "model": "agnes-2.0-flash",
    "messages": [
        {"role": "system", "content": "You are a helpful AI assistant."},
        {"role": "user", "content": "Explain how autonomous agents use tools to complete tasks. 中文回我"}
    ],
    "temperature": 0.7,
    "max_tokens": 1024
}

response = requests.post(url, headers=headers, data=json.dumps(payload))
result = response.json()
print(result["choices"][0]["message"]["content"])
```
